```python
import pandas as pd 
```

```python
df=pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\yelp\yelp_database.csv")
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["OLF"],inplace=True)
```

```python
df.columns=df.columns.str.lower().str.replace(" ","_")
```

```python
# replacing unknown values of the organization 
df["organization"]=df["organization"].fillna("not available")
```

```python
# replacing unknown values of the state
df["state"]=df["state"].fillna(df["state"].mode()[0])
```

```python
# replacing unknown values of the city
df["city"]=df["city"].fillna(df["city"].mode()[0])
```

```python
# replacing unknown values of the street and building
df["building"]=df["building"].fillna("not available")
df["street"]=df["street"].fillna("not available")
```
